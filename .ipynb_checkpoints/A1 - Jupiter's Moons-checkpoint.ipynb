# Jupiter's Moons #

### Introduction ###

This project is part of the AST221 course at the University of Toronto. In this project, using the data of Jupiter's moons in the file "AppendixA1.dat", we determined the mass of the Jupiter. 

### Code ###

```python
# Import some important libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from astropy import units as u
from scipy import stats
```

```python
# Read in the data from AppendixA1. We want data on the period and the semimajor axis of the satellites

period, semimajor = np.loadtxt('AppendixA1.dat',usecols = (5,6), skiprows = 4, unpack = True)
```

```python
# It is important to remember that the period is in days and the semimajor is in 10³ km
# To make things easier, change the period into years and the semimajor into AU

period_years = period / 365.25 * u.year
semimajor_au = (semimajor * 10 ** 3 * u.km).to(u.au)
```

```python
# Next, create the plot. In the x-axis, plot the period_years cubed and in the y-axis, plot the semimajor_au squared
# Then, add title, x-axis label, and y-axis label

fig = plt.figure()

x_axis = (period_years ** 2).value
y_axis = (semimajor_au ** 3).value

plt.scatter(x_axis, y_axis, color = 'brown')
plt.plot(x_axis, y_axis, color = 'orange')

plt.title('The Plot of Semimajor Cubed vs Period Squared of Jupiter\'s Moons')
plt.xlabel('Squared Period (yr²)')
plt.ylabel('Cubed Semimajor (au³)')

plt.show()
```

```python
# Fit the best line. Get the slope and the intercept

fit = stats.linregress(x_axis, y_axis)
slope = round(fit[0], 3)
intercept = round(fit[1], 3)

# Add the best fit line to the plot

fig = plt.figure()

x_axis = (period_years ** 2).value
y_axis = (semimajor_au ** 3).value

plt.scatter(x_axis, y_axis, color = 'brown')
plt.plot(x_axis, y_axis, color = 'orange')
plt.plot(x_axis, slope * x_axis + intercept, color = 'brown', linestyle = '--', label = 'Best fit line')

plt.xlabel('Squared Period (yr²)')
plt.ylabel('Cubed Semimajor (au³)')
plt.legend()

plt.savefig('Jupiter Moons.png')
```

### Conclusion ###

Use the Modified Kepler's Third Law:

\begin{align*}
\frac{a^3}{P^2} = M
\end{align*}

where $a$ is the orbital radius in AU, $P$ is the orbital period in years, and $M$ is mass in solar masses.

From our graph, the slope of the graph is equal to $\frac{a^3}{P^2} = M$. By fitting the data, we get the slope $\approx$ 0.00095. Since we are plotting the system of Jupiter and its moons, the slope here is equal to the mass of Jupiter in solar masses. Thus, the mass of Jupiter is around 0.00095 solar masses.

Since $1 \text{ solar mass} = 1.9891 \times 10^{30} \text{ kg}$, the mass of Jupiter is also equal to $0.00095 \times 1.9891 \times 10^{30} = 1.89 \times 10^{27} \text{kg}$.
